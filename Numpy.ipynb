```python
# Copy of an array vs an array pointing to a different location with the help of a new variable

a = np.array(l)
a

# Create a new variable and assign it a
c = a
c

# Now update 0th element of c

c[0] = 377

# This change will be reflected in c and a as both are pointing to the same location in the memory

a, c    # This is called a shallow copy
```

```python
# Now contrary to shallow copy is an actual copy of an array using np.copy()

d = np.copy(a)
d             # This is not pointing to the same location as of 'a' in memory instead a copy of a is made in different location in memory and saved in variable 'd'

a[1] = 878

a, c, d       # The output shows that the above change was made in 'a' and 'c' but not made in 'd'
```

```python
# Check the dimensions of an array

print('arr = ', arr)
print('Dimension of arr = ', arr.ndim)

array = np.array([[1, 2, 99], [3, 4, 4]])
print('array = ', array)
print('Dimension of array = ', array.ndim)

# Check the size of an array i.e. total no. of elements in an array

print('Size of array = ', array.size)

# Check the shape of an array i.e. no. of rows and colums in an array

print('Total no. rows and colums (size of an array) in array = ', array.shape)      # First number = no. of rows and second number = no. of columns

# Check the type of an array

print('Data type of array = ', array.dtype)
```

```python
# Range in NumPy

a = np.arange(1, 10)        # arange() function works same as range() function in python does
print(a)                           # a is a 1-D array
print(type(a))

# Convert 'a' (1-D array/ vector) into a 2-D array (matrix) using .reshape(rows, cols)

a = np.arange(1, 10).reshape(3, 3)      # 1st number in reshape tells the number of rows and 2nd number tell the number of cols
print('After converting a into a matrix / 2-D array: ', a)
```

```python
# Arithmetic Operations on numpy arrays

b = np.arange(1, 7).reshape(2, 3)
c = np.arange(7, 13).reshape(2, 3)
# print('a =', a, '\n b = ', b)

# Addition

print('Addition of b and c = ', b + c)
```

```python
# Convert 'a' (1-D array/ vector) into a 3-D array (tensor) using .reshape()

a = np.arange(1, 46).reshape(5, 3, 3)
a
```
